In this project I am applying 'Python for Data Science' skills to real-world financial data with some very vliable tools. First I will imort some financial data from Belex prime market (Serbian Share Market):


On the prime listing there are 3 companies:
1. NIIS - NIS a.d., Novi Sad
2. AERO - Aerodrom Nikola Tesla a.d. , Beograd
3. ENHL - Energoprojekt holding a.d. , Beograd
First we are going to import data into the data frame. The CSV files are avaliable in the folder and can be downloaded from https://www.belex.rs/eng/ with data till 31.12.2017.


```python
import pandas as pd
import matplotlib.pyplot as plt
#List of prime listing shares symbols
m_symbol=['NIIS', 'AERO', 'ENHL']
#File names of csv data
files=['belex-arhiva-NIIS.csv', 'belex-arhiva-AERO.csv', 'belex-arhiva-ENHL.csv']
listings=[]
t=0
# Import the data
for file in files:
    listing = pd.read_csv(file, sep=';', na_values=['n/a','-',''], parse_dates=['Datum trgovanja'], thousands='.', decimal=',')
    listing['Symbol'] = m_symbol[t]
    t=t+1
    listings.append(listing)

# Concatenate the listings: listing_data
listing_data = pd.concat(listings, sort=False)
```

```python
grouped=listing_data.groupby('Symbol')
plt.figure(figsize=(20,10))
for name, data in grouped:
    plt.plot(data['Datum trgovanja'].values, data['Cena'].values, 'o', label = name, alpha=0.4)
plt.legend()
plt.title('Belex Prime Market Share Prices Through Time', fontsize=18)
plt.ylabel('Price RSD')
plt.xlabel('Period')


```

```python
#Plotting with subplots, for each Company from prime listing

price=listing_data[['Datum trgovanja','Cena','Symbol']].pivot_table(values='Cena', index='Datum trgovanja', columns='Symbol', aggfunc='first')
price.plot(subplots=True, title='Price per Years')
plt.xlabel('Date')
plt.tight_layout()
plt.show()
```

```python
print(price.mean())
print(price.median())
print(price.mode())
print(price.describe())
```

```python
#Displays the distribution of three prime listing shares prices with Seaborn
#Removed missing values
import seaborn as sns
sns.set(rc={'figure.figsize':(20,8)})

price.dropna(inplace=True)
print(price.describe())

ax=sns.distplot(price['NIIS'], color='blue')
ax.axvline(price['NIIS'].mean(), color='blue', ls='--')
ax=sns.distplot(price['AERO'], color='green')
ax.axvline(price['AERO'].mean(), color='green', ls='--')
ax=sns.distplot(price['ENHL'], color='red')
ax.axvline(price['ENHL'].mean(), color='red', ls='--')
ax.set(xlabel='Price')
```

```python
#Box plot prices by prime listing shares. Many informative data in one plot
price1=listing_data[['Datum trgovanja','Cena','Symbol']].dropna()
sns.boxplot(x='Symbol', y='Cena', data=price1)
```

```python
#Swarm plot, nice variation of boxplot with informative caracteristics but you need to have limited number of opcervation
sns.swarmplot(x='Symbol', y='Cena', data=price1)
```

```python
#Change hist_symbol to show different prime listing share distributions (choose NIIS, AERO or ENHL)
hist_symbol='ENHL' # here you change the symbol
price_nis=listing_data[listing_data['Symbol']==hist_symbol].Cena.pct_change()*100
price_nis=price_nis.dropna()
plt.hist(price_nis, bins=100)
plt.show()
```

```python
# It is more usefull to compare all three distributin on subplots
import numpy as np
bins = np.linspace(-10, 10, 100)
fig, ax = plt.subplots(3, sharex=True)
t=0
color=['b','g','r']
for hysto_symbol in m_symbol:
    price_nis=listing_data[listing_data['Symbol']==hysto_symbol].Cena.pct_change()*100
    price_nis=price_nis.dropna()
    ax[t].hist(price_nis, bins=bins, alpha=0.5, color=color[t])
    ax[t].set_title(hysto_symbol)
    t=t+1
plt.show()
```
